```python
from net import st_gcn
import torch
```

```python
dev = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
def model_the_same(net1, net2):
    assert any(
        [
            p1.data.ne(p2.data).sum() > 0
            for p1, p2 in zip(net1.parameters(), net2.parameters())
        ]
    )
    assert str(net1.state_dict()) != str(net2.state_dict())
```

```python

model = st_gcn.Model(
    in_channels=3,
    num_class=60,
    dropout=0.5,
    edge_importance_weighting=True,
    graph_args={"layout": "ntu-rgb+d", "strategy": "spatial"},
).to(dev)

model2 = st_gcn.Model(
    in_channels=3,
    num_class=60,
    dropout=0.5,
    edge_importance_weighting=True,
    graph_args={"layout": "ntu-rgb+d", "strategy": "spatial"},
).to(dev)
```
